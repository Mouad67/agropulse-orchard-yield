# 01 â€“ EDA: Moroccan Orchard Yield (Synthetic)

Explore the dataset engineered to mimic Moroccan orchard conditions (avocado, apple, kaki).

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
data_path = Path('../data/moroccan_orchard_yield_synthetic.csv')
df = pd.read_csv(data_path)
df.head()
```

## Target distribution

```
plt.figure()
df['yield_kg_per_ha'].hist(bins=40)
plt.title('Yield (kg/ha) distribution')
plt.xlabel('kg/ha')
plt.ylabel('count')
plt.show()
```

## Basic correlations (numeric only)

```
num_cols = df.select_dtypes(include=[np.number]).columns
corr = df[num_cols].corr()
plt.figure(figsize=(8,6))
plt.imshow(corr, aspect='auto')
plt.colorbar()
plt.title('Correlation matrix (numeric features)')
plt.xticks(range(len(num_cols)), num_cols, rotation=90)
plt.yticks(range(len(num_cols)), num_cols)
plt.tight_layout()
plt.show()
```

## Yield by crop

```
grouped = df.groupby('crop')['yield_kg_per_ha'].describe()
grouped
```
